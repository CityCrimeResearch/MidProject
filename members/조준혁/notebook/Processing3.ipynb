### 전처리

```python
registerd_df = pd.read_csv("data/processed/등록외국인_현황_국적별_구별_2023년3분기.csv", index_col=0)
registerd_df.head()
```

```python

# 국적 + 성별 결합
registerd_df['국적_성별'] = registerd_df['국적'].astype(str).str.strip() + "(" + registerd_df['성별'].astype(str).str.strip() + ")"

# 지역 컬럼만 추출
region_columns = registerd_df.columns.difference(['국적', '성별', '소계', '국적_성별']).tolist()

# 필요한 컬럼만 남기고 transpose
region_df = registerd_df[['국적_성별'] + region_columns].copy()
region_df = region_df.set_index('국적_성별').T

# 튜플 형태 컬럼 이름 제거
region_df.columns = [col[0] if isinstance(col, tuple) else col for col in region_df.columns]

# 숫자 변환
region_df = region_df.apply(pd.to_numeric, errors='coerce')

#  확인
region_df.head()
```

```python
region_df.isna().sum()
region_df = region_df.fillna(0)
```

```python
print(region_df.isna().sum().sum()) 
```

```python
region_df.to_csv("data/processed/등록외국인_현황_국적별_구별_2023_3분기.csv", index=True, encoding='utf-8-sig')
```

### 확인용

```python
fly_df = pd.read_csv('data/processed/청소년비행데이터_2023.csv')
```

```python
fly_df
```

```python
# 접수년월 컬럼을 삭제
fly_df.drop('접수년월', axis=1, inplace=True)
fly_df.head()
```

```python
# 경찰서명이 담길 리스트를 만들어준다. 
police_name = []

for name in fly_df['경찰서명']:
    police_name.append(name)
```

```python
print('\n'.join(police_name))
```

```python
# 1대1대응 딕셔너리 만들기기
police_to_district = {
    '서울광진': '광진구',
    '서울수서': '강남구',
    '서울송파': '송파구',
    '서울서초': '서초구',
    '서울도봉': '도봉구',
    '서울동작': '동작구',
    '서울노원': '노원구',
    '서울서부': '은평구',
    '서울성동': '성동구',
    '서울강남': '강남구',
    '서울동대문': '동대문구',
    '서울종암': '성북구',
    '서울영등포': '영등포구',
    '서울서대문': '서대문구',
    '서울금천': '금천구',
    '서울강북': '강북구',
    '서울강서': '강서구',
    '서울중랑': '중랑구',
    '서울강동': '강동구',
    '서울혜화': '종로구',
    '서울양천': '양천구',
    '서울성북': '성북구',
    '서울관악': '관악구',
    '서울구로': '구로구',
    '서울중부': '중구',
    '서울은평': '은평구',
    '서울방배': '서초구',
    '서울남대문': '중구',
    '서울마포': '마포구',
    '서울종로': '종로구',
    '서울용산': '용산구',
    'ETC': '기타'
}
fly_df['자치구'] = fly_df['경찰서명'].map(police_to_district)
```

```python
fly_df.head()
```

```python
fly_grouped_df = fly_df.groupby('자치구', as_index=False).sum()
```

```python
fly_grouped_df.head()
```

```python
# 이제 경찰서명은 필요없으니 버린다
fly_grouped_df.drop('경찰서명', axis=1, inplace=True)
```

```python
# 자치구를 인덱스로
fly_grouped_df.set_index('자치구', inplace=True)
```
