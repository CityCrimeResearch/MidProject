```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

# 그래프 기본 테마 설정
sns.set()

# 폰트 파일 경로 지정
font_path = 'NanumGothic.ttf'

# 폰트 등록 및 matplotlib 설정
font_name = fm.FontProperties(fname=font_path).get_name()
fm.fontManager.addfont(font_path)
mpl.rc('font', family=font_name)
plt.rcParams['figure.figsize'] = 12, 6
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False
```

```python
df_age = pd.read_csv('1인가구(연령별)_20250602230532.csv')
df_age.head()
```

```python
df_age = df_age.iloc[1:,[1,3,4,5,6,7]]
```

```python
df_age = df_age.iloc[1:]
```

```python
df_age.to_csv('연도별_1인가구.csv', encoding='utf-8')
```

```python
df_age_sns = df_age.copy()
df_age.columns = df_age.columns.astype(str)
```

```python
df_age['2019'] = pd.to_numeric(df_age['2019'], errors='coerce')
df_age['2020'] = pd.to_numeric(df_age['2020'], errors='coerce')
df_age['2021'] = pd.to_numeric(df_age['2021'], errors='coerce')
df_age['2022'] = pd.to_numeric(df_age['2022'], errors='coerce')
df_age['2023'] = pd.to_numeric(df_age['2023'], errors='coerce')
df_age
```

```python
df_age.info()
```

```python
sns.lineplot(data=
    df_age.melt(id_vars='자치구별(2)', var_name='연도', value_name='1인가구'),
    x='연도', y='1인가구', hue='자치구별(2)')
plt.show()
```
