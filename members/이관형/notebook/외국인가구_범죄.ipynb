```python
import pandas as pd
```

```python
df_ts = pd.read_csv('transposed_seoul_df.csv')
df_ts.sort_values(by='Unnamed: 0', inplace=True, ascending=True)
df_ts.reset_index(drop=True, inplace=True)
df_ts.head()
```

```python
df_hi = pd.read_csv('주택_인구.csv')
df_hi.head()
```

```python
df_hi.sort_values(by='행정구역별(읍면동)', ascending=True, inplace=True)
df_hi.reset_index(drop=True, inplace=True)
df_hi.head()
```

```python
df_fo = pd.read_csv('외국인_인구.csv')
df_fo.head()
```

```python
df_fo.sort_values(by='행정구', ascending=True, inplace=True)
df_fo.reset_index(drop=True, inplace=True)
df_fo.head()
```

```python
df_fo.replace('-', 0, inplace=True)
df_fo.info()
df_fo.head()
```

```python
df1 = pd.DataFrame()
df1['행정구'] = df_fo['행정구']
df1['한국계 중국인'] = df_fo['한국계 중국인 남자'] + df_fo['한국계 중국인 여자']
df1['중국인'] = df_fo['중국 남자'] + df_fo['중국 여자']
df1['중국계'] = df1['한국계 중국인'] + df1['중국인']
df1['외국인'] = df_fo.iloc[:, 2:].astype(int).sum(axis=1)
df1['외국인1'] = df_hi['외국인-계 (명)']
df1['비중국계'] = df1['외국인'] - df1['중국계']
df1['인구'] = df_hi['총인구 (명)']
df1['가구'] = df_hi['가구-계 (가구)']
df1['외국인가구'] = df_hi['외국인가구 (가구)']
df1['범죄건수'] = df_ts['총 범죄 발생 건수']
df1.head()

```

```python
df1.to_csv('외국인_범죄.csv', encoding='utf-8')
```
