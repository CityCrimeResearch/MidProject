```python
import pandas as pd
```

```python
df = pd.read_csv('서울시_안심귀갓길_안전시설물.csv', encoding='cp949')
df.columns
```

```python
df1 = df[['시군구명', '시설코드','조성년월','설치대수','데이터 기준일자']]
df1
```

```python
df2 = df1.copy()
df2['시군구명'] = df2['시군구명'].str.replace(r'^(서울특별시)\s+', '', regex=True)
df2
```

```python
df2.to_csv('안심귀갓길_안전시설물.csv',encoding='utf-8')
```
