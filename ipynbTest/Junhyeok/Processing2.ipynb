```python
# 2번째 CSV 불러와서 확인
careSalary_df = pd.read_csv("data/raw/건강보험+급여(구별)_20250529152628.csv", skiprows = 2)
careSalary_df = careSalary_df.drop(careSalary_df.columns[0], axis=1)
careSalary_df
```

```python
# 컬럼명변경
careSalary_df = careSalary_df.rename(columns={'지치구별(2)': '자치구별'})
careSalary_df
```

```python
# 컬럼을 확인한다. 
for i in careSalary_df.columns:
    print(i)

# 컬럼의 개수를 파악한다. 
print("컬럼 개수:", len(careSalary_df.columns))
```

```python
careSalary_df.info()
```

```python
careSalary_df.dtypes
```

```python
# 소계, 지치구별 열을 다 없애버린다. 
careSalary_df = careSalary_df[
    (careSalary_df['자치구별'] != '소계') & (careSalary_df['자치구별'] != '지치구별(2)')
]
careSalary_df
```

```python
# 인덱스를 리셋해준다
careSalary_df = careSalary_df.reset_index(drop=True)
careSalary_df
```

```python
# 컬럼 이름을 바꿔준다. 
careSalary_df.columns = [col.replace('.1', '(건수)') for col in careSalary_df.columns]
careSalary_df.columns = [col.replace('(1)', '') for col in careSalary_df.columns]
careSalary_df
```

```python
# csv로 내보낸다.
careSalary_df.to_csv('data/processed/건강보험_급여구별여구별.csv', encoding='utf-8-sig', index=True)
```
