```python
# 저장된 파일을 읽어온다다
crime_df = pd.read_csv('data/seoul_raw/경찰청_범죄 발생 지역별 통계_20231231.csv', encoding='cp949')
crime_df.set_index('범죄중분류', inplace=True)
crime_df
```

```python
seoul_crime_df = crime_df.loc[:, crime_df.columns.str.contains('서울')]
seoul_crime_df
```

```python
# 컬럼을 확인한다. 
for i in seoul_crime_df.columns:
    print(i)

# 컬럼의 개수를 파악한다. 
# 25개가 나와야함
print("컬럼 개수:", len(seoul_crime_df.columns))
```

```python
# 서울이라는 단어가 앞에 붙은 컬럼명을 제거한 새 이름으로 바꿔준다. 
seoul_crime_df.columns = seoul_crime_df.columns.str.replace('^서울', '', regex=True)
seoul_crime_df
```

```python
# 인덱스와 컬럼을 바꾼다. 
transposed_seoul_df = seoul_crime_df.transpose()
# 인덱스의 이름을 바꾼다. 
transposed_seoul_df.index.name = None

transposed_seoul_df
```

```python
# 총 범죄 발생 건수라는 컬럼을 새로 만든다. 
transposed_seoul_df['총 범죄 발생 건수'] = transposed_seoul_df.sum(axis=1)
# 확인한다. 
transposed_seoul_df
```

```python
# csv로 내보낸다.
transposed_seoul_df.to_csv('data/processed/transposed_seoul_df.csv', encoding='utf-8-sig', index=True)
```
